# *Exercise 6: Built-in Functions*

##*Write a program that:*
##*Takes a list of numbers as input from the user.*
##*Uses the abs(), max(), and min() functions to find the absolute values,*
##*maximum, and minimum of the numbers.*
##*Prints the results with appropriate messages*


```python
# Taking input from the user

user_input = input("Enter a list of numbers separated by commas")

# Converting the input string into a list of numbers
number_list = [float(num) for num in user_input.split(",")]

# Using built-in functions to find absolute values, maximum, and minimum
absolute_values = [abs(num) for num in number_list]  # Absolute values
max_value = max(number_list)                           # Maximum value
min_value = min(number_list)                           # Minimum value

# Printing the results
print("Original list of numbers:", number_list)
print("Absolute values:", absolute_values)
print("Maximum value:", max_value)
print("Minimum value:", min_value)
```

Explanation:

Input: The program prompts the user to enter a list of numbers separated by commas.

Conversion: It splits the input string into individual number strings, converts them to floats, and stores them in a list called number_list.

Calculations:
Absolute Values: A list comprehension is used to create a list of absolute values.

Maximum Value: The max() function finds the maximum value in the list.

Minimum Value: The min() function finds the minimum value in the list.


Explanation:

This program effectively demonstrates the use of built-in functions and provides a clear output based on user input.

# *Exercise 7: Arithmetic Operations*

## *Create a simple calculator program that:*
## *Prompts the user to enter two numbers.*
## *Asks the user to select an operation: addition, subtraction,*
## *multiplication, or division.*
## *Performs the selected operation and prints the result. Handle division by zero gracefully.*


```python
# Prompting the user to enter two numbers

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Asking the user to select an operation

operation = input("Select an operation (add, subtract, multiply, divide): ").lower()

# Performing the selected operation

if operation == "add":
    result = num1 + num2
    print("Result:", result)
elif operation == "subtract":
    result = num1 - num2
    print("Result:", result)
elif operation == "multiply":
    result = num1 * num2
    print("Result:", result)
elif operation == "divide":
    if num2 != 0:  # Handling division by zero
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation selected.")
```

Explanation:

Input: The program prompts the user to enter two numbers and an operation.

Operation Selection: It checks the selected operation using conditional statements.

Calculation: It performs the chosen operation:

For addition, subtraction, and multiplication, it calculates and prints the result.

For division, it checks if the second number is zero to avoid division by zero, printing an error message if it is.

Error Handling: It includes a message for invalid operations
